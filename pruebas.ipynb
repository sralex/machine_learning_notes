```python
import pandas as pd

# Crear un DataFrame de ejemplo
data = {
    'Producto': ['A', 'B', 'A', 'C', 'B', 'A'],
    'Región': ['Norte', 'Sur', 'Norte', 'Sur', 'Norte', 'Sur'],
    'Ventas': [100, 150, 200, 50, 300, 75]
}
df = pd.DataFrame(data)

# Agrupar por producto y calcular la suma de ventas
df.groupby('Producto')[['Ventas']].sum().reset_index()
```

```python
# Agrupar por región y calcular la media de ventas
df.groupby('Región')['Ventas'].mean()
```

```python
df.groupby(['Producto', 'Región'])['Ventas'].mean()
```

```python
# Función personalizada para calcular el rango
def rango(x):
    return x.max() - x.min()

df.groupby('Producto')['Ventas'].apply(lambda x: x.max()- x.min())
```

```python
df.groupby('Producto')['Ventas'].agg(['sum', 'mean'])
```

```python
data = {
    'Producto': ['A', 'B', 'A', 'C', 'B', 'A'],
    'Región': ['Norte', 'Sur', 'Norte', 'Sur', 'Norte', 'Sur'],
    'Ventas': [100, 150, 200, 50, 300, 75],
    'Cantidad': [10, 20, 15, 5, 30, 8]
}
df = pd.DataFrame(data)

```

```python
# Agrupar por región y aplicar múltiples funciones a diferentes columnas
df.groupby('Región').agg({
    'Ventas': ['sum', 'mean'],
    'Cantidad': 'count'
})
```

```python
df.groupby('Producto')['Ventas'].sum()
```

```python
df['ventas_producto'] = df.groupby('Producto')['Ventas'].transform(lambda x: x.sum())
df
```

```python
df.groupby('Producto').filter(lambda x: x['Ventas'].sum() > 200)
```
